Q9) The column 'Size' contains alphanumeric values, treat the non numeric data and convert the column into suitable data type.

```python
data = appsdata.copy()
data
```

```python
len(data[(data['Size']=='Varies with device')])
```

```python
#1 dropping rows with size as 'Varies with device'

data.drop(data[data['Size']=='Varies with device'].index, inplace = True)
```

```python
len(data[(data['Size']=='Varies with device')])
```

```python
#2 converting M to 1000000 and k to 1000
num = {'M':1000000, 'k':1000}

data['Size'] = (data['Size'].str[0:-1].astype(float)*data['Size'].str[-1].replace(num)).astype(float)
```

```python
data
```

Q13) Encode the categorical columns.

#To encode the categorical columns, we are using Label-encoder method as the number of unique values in each feature is a significantly big number and the data would become too big and tough to understand and interpret.

```python
encoded = LabelEncoder()
for col in data1.columns:
    if data1[col].dtype=='object':
        data1[col]=encoded.fit_transform(data1[col])
```
